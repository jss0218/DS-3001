# Simple Cleaning

```python
# For city_records, clean the 'income' and 'tax_due' columns by removing "$" and "," and convert to float.

df_city['income_clean'] = (
    df_city['income']
    .replace({'\$': '', ',': ''}, regex=True)
    .astype(float)
)
df_city['tax_due_clean'] = (
    df_city['tax_due']
    .replace({'\$': '', ',': ''}, regex=True)
    .astype(float)
)

df_city.head()
```

Calling GroupBy creates a **GroupBy object**.

It's like a dataframe separated out by each unique value of the column we specified (content_rating).

We can apply various operations to it and get **mean, min, etc. values for each content rating**, instead of for the whole column.

```python
# Calling GroupBy
df_city.groupby('housing_status_clean')
```

Important to note here is that calls of these aggregation functions (min, max, mean, or any custom-defined functions) on groupby objects return the following object:

 - A **Series**, if only one function is applied to the groups.
 - A **DataFrame**, if multiple functions are applied.


This should make sense.


**In either case**, the index of the series are the group names.

 This allows you to easily grab out specific groups, like this:


```python
# Calling GroupBy with Mean
grouped_housing = df_city.groupby('housing_status_clean')['income_clean'].mean()
grouped_housing
```

```python
gb1 = grouped_housing.reset_index()
gb1
```

```python
gb1 = gb1.rename(columns={'income_clean': 'avg_income', 'housing_status_clean': 'housing_status'})
gb1
```

## Using GroupBy Aggregation

```python
agg_result = df_city.groupby('housing_status_clean').agg({
    'income_clean': ['mean', 'sum', 'min', 'max'],
    'tax_due_clean': ['mean', 'sum']
})
agg_result
```

Aggregation by multiple columns often gives you multiple levels of headers. Calling reset_index() (even multiple times) will not solve this because the header structure has been dramatically changed.

```python
agg_result.reset_index()
```

```python
agg_result.reset_index().reset_index()
```

Instead, use droplevel() to get rid of the extra level, and then call reset_index() as necessary. You might have to do this multiple times depending on how many header rows you end up with!

```python
agg_result.columns = agg_result.columns.droplevel(0)
agg_result
```

For best practice, be sure to rename your columns directly if you drop headers. Otherwise, your column names can get very confusing if they are named generic things like "mean" or "max".

```python
agg_result.reset_index()
```

```python
agg_result.columns
```

You should rename columns before resetting index, otherwise you may run into Index Errors.

```python
agg_result.columns = ['Income Mean', 'Income Sum', 'Income Min', 'Income Max',  'Tax Due Mean', 'Tax Due Sum']
agg_result
```

```python
agg_result.reset_index()
```

```python
grouped_city = df_city.groupby('housing_status_clean').agg(
    income_mean=('income_clean', 'mean'),
    income_sum=('income_clean', 'sum'),
    tax_mean=('tax_due_clean', 'mean'),
    tax_sum=('tax_due_clean', 'sum')).reset_index()
grouped_city
```

## Aggregation With Custom Functions

```python
df_city
```

You are limited to using .agg() with the functions supported by Pandas.

```python
df_city['housing_status_clean'].agg(['mode'])
```

....unless you make your own!

```python
# Define a custom aggregation function to compute the mode:
def custom_mode(series):
    modes = series.mode()
    # Return the first mode if available; else return NaN
    return modes.iloc[0] if not modes.empty else np.nan
```

```python
agg_result = df_city['housing_status_clean'].agg(mode_value=custom_mode)
```

```python
agg_result
```

## Plotting GroupBy Analyses

```python
# 1. Bar Plot: Average Income by Housing Status (from df_city)
plt.figure(figsize=(8, 6))
plt.bar(grouped_city['housing_status_clean'], grouped_city['income_mean'], color='skyblue')
plt.xlabel('Housing Status')
plt.ylabel('Average Income')
plt.title('Average Income by Housing Status')
plt.show()
```

```python
# 2. Box Plot: Monthly Expenses Distribution by Food Assistance Program (from df_food)
plt.figure(figsize=(8, 6))
sns.boxplot(x='program_assigned_clean', y='monthly_expenses_clean', data=df_food, palette="Set2")
plt.xlabel('Program Assigned')
plt.ylabel('Monthly Expenses')
plt.title('Monthly Expenses Distribution by Food Assistance Program')
plt.show()
```

```python
# 3. Error Bar Plot: Average Monthly Expense by Program with Standard Deviation
# First, group by program_assigned_clean for overall program-level aggregation.
program_avg = df_food.groupby('program_assigned_clean').agg(
    avg_monthly_expense=('monthly_expenses_clean', 'mean'),
    expense_std=('monthly_expenses_clean', 'std')
).reset_index()
program_avg
```

```python
plt.figure(figsize=(8, 6))
plt.errorbar(
    program_avg['program_assigned_clean'],
    program_avg['avg_monthly_expense'],
    yerr=program_avg['expense_std'],
    fmt='o',
    capsize=5,
    color='coral'
)
plt.xlabel('Program Assigned')
plt.ylabel('Average Monthly Expense')
plt.title('Average Monthly Expense by Program (with STD)')
plt.show()
```
